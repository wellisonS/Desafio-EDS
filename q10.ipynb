```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def visualizar_atendimentos(lista_datas):
    # Contar a frequência de cada data
    contador_datas = Counter(lista_datas)

    # Separar as datas e as contagens em listas distintas
    datas = list(contador_datas.keys())
    contagens = list(contador_datas.values())

    # Configurar o gráfico de barras
    plt.bar(datas, contagens)
    plt.xlabel('Data')
    plt.ylabel('Quantidade de Atendimentos')
    plt.title('Quantidade de Atendimentos Médicos por Dia')

    # Ajustar o espaçamento entre as barras no eixo x
    plt.xticks(rotation=45)

    # Exibir o gráfico
    plt.show()
```

```python
lista_datas = ['2023-05-01', '2023-05-01', '2023-05-02', '2023-05-03', '2023-05-03', '2023-05-03']
visualizar_atendimentos(lista_datas)
```

```python
lista_datas = ['2023-05-11', '2023-05-11', '2023-05-11', '2023-05-11', '2023-05-03', '2023-05-03']
visualizar_atendimentos(lista_datas)
```
